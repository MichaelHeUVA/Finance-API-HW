# Task 2.1


```python
stock21 = input("Enter the stock symbol: ")
```

```python
url21 = f"https://yfapi.net/v11/finance/quoteSummary/{stock21}?modules=%20summaryDetail%2CfinancialData"


response21 = requests.request("GET", url21, headers=headers)
```

```python
print(json.dumps(response21.json(), indent=2))
```

```python
df = pd.DataFrame(response21.json()["quoteSummary"]["result"][0])
df
```

```python
while True:
    option = input(
        "1. Find 52 Week High\n2. Find the 52 Week Low\n3. Find the Return on Assets (ROA), 4. Exit\n"
    )
    if option == "1":
        df_data1 = {
            "Ticker": [stock21.upper()],
            "52 Week High": [f"${df['summaryDetail']['fiftyTwoWeekHigh']['fmt']}"],
        }
        df1 = pd.DataFrame(df_data1)
        print(df1)
    elif option == "2":
        df_data2 = {
            "Ticker": [stock21.upper()],
            "52 Week Low": [f"${df['summaryDetail']['fiftyTwoWeekLow']['fmt']}"],
        }
        df2 = pd.DataFrame(df_data2)
        print(df2)
    elif option == "3":
        df_data3 = {
            "Ticker": [stock21.upper()],
            "Return on Assets": [df["financialData"]["returnOnAssets"]["fmt"]],
        }
        df3 = pd.DataFrame(df_data3)
        print(df3)
    elif option == "4":
        break
    else:
        print("Invalid option")
```
